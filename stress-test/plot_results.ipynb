```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('results.csv')
results_leaderless = pd.read_csv('results_leaderless.csv')
results
```

```python
plt.plot(range(96), results['Latency'][:96], label='Latency with leader')
plt.plot(range(96), results_leaderless['Latency'][:96], label='Latency leaderless')
plt.legend(loc='best')
plt.show()
```

```python
plt.boxplot(results['Latency'][:96])
plt.legend(['Latency with leader'], loc='best')
plt.show()
```

```python
plt.boxplot(results_leaderless['Latency'][:96])
plt.legend(['Latency leaderless'], loc='best')
```
