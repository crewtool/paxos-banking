```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('results.csv')
results_leaderless = pd.read_csv('results_leaderless.csv')
results
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(range(96), results['Latency'][:96], label='Latency with leader')
ax.plot(range(96), results_leaderless['Latency'][:96], label='Latency leaderless')
ax.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.boxplot(results['Latency'][:96])
ax.legend(['Latency with leader'], loc='best')
fig.patch.set_facecolor('white')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.boxplot(results_leaderless['Latency'][:96])
ax.legend(['Latency leaderless'], loc='best')
fig.patch.set_facecolor('white')
plt.show()
```
